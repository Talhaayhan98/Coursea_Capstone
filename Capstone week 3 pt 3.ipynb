```python
import pandas as pd
import numpy as np
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

dfs = pd.read_html(url)
                                                                                                           
df = dfs[0]
                                                                                                         
df2 = df[['Postal Code','Borough','Neighbourhood']]

Canada_Postal=pd.DataFrame(df2)
Canada_Postal

# select all the Boroughs that are not (not assignmed), quick easy way to remove these values
df = Canada_Postal[Canada_Postal.Borough != 'Not assigned']
df=df.rename(columns={'Postal Code':'PostalCode'})

# upload latitude longitude
#call the cvf file that has the postal code and latitude longitude
url='http://cocl.us/Geospatial_data'
lat_lan=pd.read_csv(url)
#convert to pandas data frame
lat_lan=pd.DataFrame(lat_lan)
lat_lan.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df_main1=df.merge(lat_lan)
df_main1
```

```python
from geopy.geocoders import Nominatim
address = 'Toronto'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude,longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighbourhood in zip(df_main1['Latitude'], df_main1['Longitude'], df_main1['Borough'], df_main1['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_toronto)  
    
map_toronto
```
